```python
unique, counts = np.unique(sub['PROVINCE_CODE'], return_counts=True)
unique_count_province = dict(zip(unique, counts))
unique_count_province
```

**Chia các thành phố ra 3 nhóm: Lớn, Vừa và Nhỏ**

- Lớn gồm các thành phố có trên 35 thuê bao (DTH, HCM, HNO, HPH, QNI, THO, VLO)
- Vừa gồm các thành phố có từ 15 đến 35 thuê bao (AGI,BGI,CTH, DNA, DNI, HDU, HGI, HTI, KHO, NAN, QBI, QNA, TBI, TGI, TTH, TVI)
- Nhỏ gồm các thành phố có dưới 15 thuê bao (các tỉnh còn lại)

```python
large_cities = []
med_cities = []
small_cities = []

for key, value in unique_count_province.items():
    if value >= 35:
        large_cities.append(key)
    elif 15 <= value and value < 35:
        med_cities.append(key)
    else:
        small_cities.append(key)

print(large_cities)
print(med_cities)
print(small_cities)
```

```python
for i in range(len(sub)) : 
    if sub.loc[i, "PROVINCE_CODE"] in large_cities:
        sub.loc[i, "PROVINCE_CODE"] = "Large City"
    elif sub.loc[i, "PROVINCE_CODE"] in med_cities:
        sub.loc[i, "PROVINCE_CODE"] = "Medium City"
    else:
        sub.loc[i, "PROVINCE_CODE"] = "Small City"
```

## PCA

I will try to reduce the dimension of the dataset but at them same time keep the highest accuracy possible

```python
from sklearn.decomposition import PCA

pca_99per = PCA()
pca_99per.fit(sub)
cumsum = np.cumsum(pca_99per.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.999999) + 1

d
```

We only need 1 to achive almost 100% variance

```python
plt.figure(figsize=(6,4))
plt.plot(cumsum, linewidth=3)
plt.axis([0, 8, 0, 1])
plt.xlabel("Dimensions")
plt.ylabel("Explained Variance")
plt.plot([d, d], [0, 0.99], "k:")
plt.plot([0, d], [0.99, 0.99], "k:")
plt.plot(d, 0.99, "ko")

plt.grid(True)
save_fig("explained_variance_plot")
plt.show()
```

Not really helpful. Let's just reduce the dimension down to 2.

```python
pca_2d = PCA(n_components = 2)

sub_pca_reduced_2d = pca_2d.fit_transform(sub)

sub_pca_reduced_2d.shape
```
